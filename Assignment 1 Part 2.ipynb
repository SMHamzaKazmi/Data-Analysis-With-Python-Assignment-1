```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("clean_chicago_traffic.csv")
```

```python
df
```

## Graph 1

```python
graph = df['ROADWAY_SURFACE_COND'].value_counts()

# Plot a bar graph
plt.figure(figsize=(12, 8))  # Adjust figure size for horizontal bars
sns.barplot(x=graph.values, y=graph.index, orient='h')
plt.title('Frequency of Accidents by Roadway Surfare Condition')
plt.xlabel('Number of Accidents')
plt.ylabel('Roadway Surface Condition')
plt.tight_layout()
plt.show()
```

## Graph 2

```python
graph = df['CRASH_MONTH'].value_counts().sort_index()

# Plot a line graph
plt.figure(figsize=(12, 6))
graph.plot(kind='line', marker='o', color='red')
plt.title('Monthly Trend of Crashes')
plt.xlabel('Month')
plt.ylabel('Number of Crashes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Graph 3

```python
graph = df['CRASH_DAY_OF_WEEK'].value_counts()

# Plot a bar graph
plt.figure(figsize=(10, 6))
sns.barplot(x=graph.index, y=graph.values, color='green')
plt.title('Weekly Distribution of Crashes')
plt.xlabel('Days in a week')
plt.ylabel('Number of Crashes')
plt.tight_layout()
plt.show()
```

## Graph 4

```python
graph = df['LIGHTING_CONDITION'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(graph, labels=graph.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('flare'))
plt.title('Distribution of Lighting Conditions of Accidents')
plt.tight_layout()
plt.show()
```

## Graph 5

```python
# Plot a scatter plot for geographical distribution
plt.figure(figsize=(10, 8))
sns.scatterplot(x='LONGITUDE', y='LATITUDE', data=df, hue='TRAFFICWAY_TYPE', palette='viridis', alpha=0.6)
plt.title('Geographical Distribution of Trafficway Type of Accidents')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.ylim([41.5, 42.2])
plt.xlim([-88, -87.5])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside the plot
plt.tight_layout()
plt.show()
```
