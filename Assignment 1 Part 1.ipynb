```python
import pandas as pd
```

```python
df = pd.read_csv("Traffic_Crashes_-_Crashes.csv")
```

```python
df
```

## Cleaning data by dropping and filling NaN values

```python
# Dropping the columns that do not seem much important and have many NaN values so it is better to drop them instead of filling
df = df.drop(columns=['CRASH_DATE_EST_I', 'LANE_CNT', 'INTERSECTION_RELATED_I', 'NOT_RIGHT_OF_WAY_I', 'PHOTOS_TAKEN_I', 'STATEMENTS_TAKEN_I', 'DOORING_I', 'WORK_ZONE_I', 'WORK_ZONE_TYPE', 'WORKERS_PRESENT_I'])

df
```

```python
# HIT_AND_RUN_I is important column so not dropping it and filling NaN values
df.HIT_AND_RUN_I
```

```python
# Filling NaN values with 'UNKNOWN'
df['HIT_AND_RUN_I'] = df['HIT_AND_RUN_I'].fillna("UNKNOWN")
```

```python
# NaN values filled in HIT_AND_RUN_I
df.HIT_AND_RUN_I
```

```python
# Checking the number rows that have at least one NaN value
num_rows_with_nan = df.isna().any(axis=1).sum()
print(num_rows_with_nan)
```

```python
# Dropping the rows that have at least one NaN value instead of filling them because they are a small number as compared to the whole data
df = df.dropna()
```

```python
# Checking number of rows that have at least one NaN value
num_rows_with_nan = df.isna().any(axis=1).sum()
print(num_rows_with_nan)
```

```python
# Checking number of duplicate rows
num_duplicates = df.duplicated().sum()
print(num_duplicates)
```

## Data is cleaned and there are no NaN values
## Now handling the date time and splitting them into separate columns of date and time

```python
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'])
df
```

```python
# Splitting CRASH_DATE into separate columns of date and time
df['Crash Date'] = df['CRASH_DATE'].dt.date
df['Crash Time'] = df['CRASH_DATE'].dt.time

df
```

```python
df['DATE_POLICE_NOTIFIED'] = pd.to_datetime(df['DATE_POLICE_NOTIFIED'])
df
```

```python
# Splitting DATE_POLICE_NOTIFIED into separate columns of date and time
df['Police Notified Date'] = df['DATE_POLICE_NOTIFIED'].dt.date
df['Police Notified Time'] = df['DATE_POLICE_NOTIFIED'].dt.time

df
```

```python
# Dropping the columns that were split because they are useless now
df = df.drop(columns=['CRASH_DATE', 'DATE_POLICE_NOTIFIED'])
```

```python
# Checking the unique values of POSTED_SPEED_LIMIT as it has some outliers
# In real life speed limit is mostly between 20 and 60 so the values less than 20 and greater than 60 are outliers
unique_values = df['POSTED_SPEED_LIMIT'].unique()
sorted_unique_values = sorted(unique_values)

print(sorted_unique_values)
```

```python
# Saving to new file
df.to_csv("clean_chicago_traffic.csv", index=False)
```
